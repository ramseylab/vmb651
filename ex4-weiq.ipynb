# Import Dependecies
Biopython package

```python
from Bio import Entrez
```

# Print out number of records containing SARS-CoV-2

```python
Entrez.email = "weiq@oregonstate.edu"
result = Entrez.esearch(db="nucleotide",
                        term="SARS-CoV-2",
                        idtype="acc",
                        retmax=1000)

records = Entrez.read(result)
print("Number of records containing SARS-CoV-2 is: ", records['Count'])
#print("Number of records containing SARS-CoV-2 is: ", len(records))
#with Entrez.esearch(db="nucleotide",
#                    term="SARS-CoV-2",
#                    idtype="acc",
#                    retmax=1000) as result_handle:
```

# Print out dates and accession numbers of all COVID-19 sequences

```python
with Entrez.esearch(db="nucleotide",
                    term="SARS-CoV-2",
                    #idtype="acc",
                    retmax=60000) as result_handle:
    sequence_ids_combined = ','.join(Entrez.read(result_handle)['IdList'])
    with Entrez.efetch(db="nucleotide",
                       retmode="xml",
                       id=sequence_ids_combined) as sub_result_handle:
            for record in Entrez.read(sub_result_handle):
                    seq_date = record['GBSeq_update-date']
                    seq_accession = record['GBSeq_primary-accession']
                    print("Dates and accession numbers of COVID-19 sequences: ", seq_date, seq_accession)
```
